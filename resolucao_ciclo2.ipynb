Como é a variação do preços dos imóveis em NY?

Acessando a coluna de preço dos imoveis de NY

```python
price = dt.loc[:, 'price']
display(price)
```

Variaçao eh desvio padrao

```python
price_remove = price.dropna()
display(price_remove)
```

```python
print(price_remove.isnull().sum())
```

Usando a funçao std em numpy

```python
desvio_padrao = np.std(price_remove)
print(desvio_padrao)
```

Calculando a media de imoveis de NY

```python
media = np.mean(price_remove)
print(media)
#lista = list(price_remove)
```

Calculando o valor maximo de preços e minimo

```python
#maximo
max = np.max(price_remove)
print(f' preço maximo {max}')
#minimo
min = np.min(price_remove)
print(f'preço minimo {min}')
```

Filtrando linhas menores que 1250


```python
price_menor_1250 = dt.loc[(dt['price']< 1250)]
display(price_menor_1250)
price_coluna_1250 = price_menor_1250['price']
display(price_coluna_1250)
max = np.max(price_coluna_1250)
min = np.min(price_coluna_1250)
print(max)
print(min)
```

Existem mais imóveis baratos ou caros?

Criar um histograma o numero de alugueis caros e baratos

```python
plt.ylabel('numero de imoveis')
plt.xlabel('preço do aluguel')
plt.legend()
histograma = plt.hist(price_coluna_1250,bins=12, color = "purple");
```
